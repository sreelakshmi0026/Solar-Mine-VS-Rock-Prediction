# **Importing the dependencies**

```python
import numpy as np # NumPy is a library for numerical computing in Python.
import pandas as pd # library for data manipulation, Data Loading( like pandas can read various file formats like CSV) making it easy to load into the evnironment. Handling large datasets, Data Cleaning, Data Transformation etc...
import matplotlib.pyplot as plt # used for making Plots, Graphs, and Charts
import seaborn as sns  # used for statistical data visualization, Statistical Plots etc...
from sklearn.model_selection import train_test_split # This is used to split the data into training and testing data
from sklearn.linear_model import LogisticRegression # Thats how we import the logisticregression model
from sklearn.metrics import accuracy_score # This is used to find the accuracy of the model
```

# ***Data collection and Data Processing ***

```python
# Loading the dataset into Pandas dataframe
sonar_data = pd.read_csv('/content/Copy of sonar data.csv', header=None)
```

**MORE THE DATA MORE ACCURATE THE MODEL IS ! WE NEED TO NOTE THIS (but we have small amt of data here)**

```python
# Separating data and labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

# **Training and Testing Data**


```python
# training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_test.shape, X_train.shape) # the output it prints now is (208, 60) in the original data we have this (21,60)in the test data we have this and (187,60) in the training data we have this
```

# **Model Training ---> Logistic Regression**

```python
model = LogisticRegression()
```

```python
# Training the logistic regression model with training
model.fit(X_train, Y_train)
```

# **Model Evaluation**

```python
# Model evaluation
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
# Now lets fid the accuracy score of test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on testing data : ', testing_data_accuracy)
```
